## Data Processing Procedure

Here in this part of code, We are going to process our html files.
Each HTML file contains information about a toyota car listing.

We will extract certain variables like Vin, Mileage, Model , year etc.

The data will be first organized in a dataframe.

This dataframe will then be used for our Machine Learning Model. (Our analysts will export as a csv per requirements)

At the final step, we will use this dataframe to store the contents in local mongodb database.



```python
import os

from bs4 import BeautifulSoup

import time
import os
import pandas as pd
import numpy as np


from pymongo import MongoClient
```

-----

```python
# Directory where html files which were scraped are stored

desired_path = '/Users/jaskeeratbrar/Documents/DDR ML Final laptop/Toyota Used'

os.chdir(desired_path)

filenames = []

for filename in os.listdir(os.getcwd()):
    if filename.endswith('.html'):
        filenames.append(filename)
```

```python
# define variables of interest
info = ['Year','Make','Model','Trim','Body','Cab Type','Truck Bed Size','Exterior Condition','Interior Condition',
        'Drive Type','Dealer License','Stock Number','Transmission','Liters','Cylinders','Fuel Type','Number of Doors',
        'Number of Seat Belts','Mileage','VIN','Title Type','Exterior Color','Interior Color']


vehicle_specifications = []


for file in filenames:
    
    with open(file,'r',encoding = 'UTF-8') as file:
        car_html = file.read()

    
    soup_car = BeautifulSoup(car_html)

    car = []
    
    for i in info:
        try:
            car_info = soup_car.find('div', string = f'{i}').find_next_sibling().text
            car.append(car_info)
        except AttributeError:
            car.append('not found')

    print(file)
    vehicle_specifications.append(car)
```

```python
## inspect the variable to verify data collection/ processing is happeninng as expected

vehicle_specifications
```

```python
# transform to a data frame

vehicle_specifications_df = pd.DataFrame(vehicle_specifications, columns=info)
```

```python
# inspect df, looks like row 452 indicates presence of a bad row.
# we need to deal with this data quality issue

vehicle_specifications_df
```

----

```python
# We wiill only store vehicle info in a row if it has info about the year of vehicle
# if there is no numerical value, we drop it
# any na value gets dropped as well

vehicle_specifications_df = vehicle_specifications_df[vehicle_specifications_df['Year'] != 'not found']

vehicle_specifications_df = vehicle_specifications_df.dropna(subset=['Year'])

```

```python
# inspect data

vehicle_specifications_df
```

```python
# variables for storing below values 

price = []
mileage = []
views = []


for file in filenames:
    print(file)

    with open(file,'r',encoding = 'UTF-8') as file:
        car_html = file.read()

   
    soup_car = BeautifulSoup(car_html)
    
    try:
    
        price.append(soup_car.select('h2')[0].text)
    except IndexError:
        price.append('page disappear')
    
    try:
        # get the  mileage
        mileage.append(soup_car.select('h2')[0].find_next_sibling().text)
    except IndexError:
        mileage.append('page disappear')

    try:
        # get the number of views
        views.append(soup_car.select_one('div.ShareButtons__ViewWrapper-sc-1m1698q-1.gMqvWY>p.Typography__variantProp-sc-5cwz35-0.kMOSqH').text)
    except IndexError:
        views.append('page disappear')
    except AttributeError:
        views.append('not found')
```

```python
# convert the list to dataframe
price_df = pd.DataFrame({'Price':price})
view_df = pd.DataFrame({'Views':views})

# concatenate newly obtained df above with the dataframe
df = pd.concat([vehicle_specifications_df,price_df,view_df], axis = 1)

# delete comma in mileage
df['Mileage'] = df['Mileage'].replace(',','',regex = True)

# delete $ in price
df['Price'] = df['Price'].replace({'\$': '', ',': ''}, regex=True)

# delete views in Views
df['Views'] = df['Views'].replace(' Views','', regex = True)

```

```python
df
```

```python
## data cleaning, removing na values

# replace not found cells as na
df.replace('not found', np.nan, inplace=True)

# replace page disappear cells as na
df.replace('page disappear', np.nan, inplace=True)

```

```python
df = df.dropna(subset=['Year'])
```

```python
df
```

```python
# convert character to number

df[['Mileage','Price','Views']].astype('float')
```

-----

```python


# init connection to mongo

client = MongoClient('mongodb://localhost:27017/')

#  database
db = client['Used_Cars_Dataset']

collection = db['Toyota']

# Before converting the df to a dictionary, set the 'Vin' column as the index

# This makes 'Vin' the key in each dictionary, which we will rename to '_id'.
df.set_index('VIN', inplace=True)

# Convert the DataFrame to a list of dictionaries
data = df.reset_index().rename(columns={'Vin': '_id'}).to_dict('records')


# Insert the documents into the collection
# If a document with the same '_id' already exists, this will raise a DuplicateKeyError.
collection.insert_many(data)

sample_documents = list(collection.find().limit(5))

```

```python

# Close the MongoDB connection
client.close()
```
